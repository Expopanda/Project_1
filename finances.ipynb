```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

financial_file = Path("Resources/salaries_per_city.csv")
```

```python
financial_df = pd.read_csv(financial_file)
financial_df.head()
```

```python
#Total amount paid to the emergency services departments per city?
cities = financial_df["City"]
fire_per_city = []
police_per_city = []

for city in cities:
    fire_salaries = (financial_df.loc[financial_df["City"] == city, "Fire"].sum())/1000000
    police_salaries = (financial_df.loc[financial_df["City"] == city, "Police"].sum())/1000000
    fire_per_city.append(fire_salaries)
    police_per_city.append(police_salaries)

x_axis = np.arange(len(cities))
plt.figure(figsize=(8, 6))

plt.bar(x_axis - 0.2, fire_per_city, width=0.4, label='Fire Salaries', color='blue')
plt.bar(x_axis + 0.2, police_per_city, width=0.4, label='Police Salaries', color='green')

plt.title("Total Amounts Paid for Emergency Services per City")
plt.xlabel("City")
plt.ylabel("Amount Paid for Services (in millions)")
plt.xticks(x_axis, cities)
plt.legend() 

plt.tight_layout()
plt.show()
```

```python
#What is the average amount per person of emergency services in each city?
cities = financial_df["City"]
average_ems_per_person = []

for city in cities:
    total_ems_expenditure = financial_df.loc[financial_df["City"] == city, "Total Emergency Services"].sum()
    total_population = financial_df.loc[financial_df["City"] == city, "Population"].sum()

    if total_population != 0:
        ems_per_person = total_ems_expenditure / total_population
        average_ems_per_person.append(ems_per_person)
    else:
        average_ems_per_person.append(0)

for city, average_ems in zip(cities, average_ems_per_person):
    print(f"{city}: {average_ems}")

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.swarmplot(x="City", y=average_ems_per_person, data=financial_df)

plt.xlabel("City")
plt.ylabel("EMS Expenditure per Person")
plt.title("EMS Expenditure per Person by City")
plt.xticks(rotation=45, ha="center")
plt.show()
```
