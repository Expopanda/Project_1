```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

FBI10_path = 'Resources/FBI_2010_illinois_amended.csv'
FBI11_path = 'Resources/FBI_2011_illinois_amended.csv'
FBI12_path = 'Resources/FBI_2012_illinois_amended.csv'
FBI13_path = 'Resources/FBI_2013_illinois_amended.csv'
FBI14_path = 'Resources/FBI_2014_illinois_amended.csv'
FBI15_path = 'Resources/FBI_2015_illinois_amended.csv'
FBI16_path = 'Resources/FBI_2016_illinois_amended.csv'
FBI17_path = 'Resources/FBI_2017_illinois_amended.csv'
FBI18_path = 'Resources/FBI_2018_illinois_amended.csv'
FBI19_path = 'Resources/FBI_2019_illinois_amended.csv'
```

```python
# these are the dataframes with all NaN values in tact. Please don't amend them. 

crime10_base = pd.read_csv(FBI10_path, delimiter=',')
crime11_base = pd.read_csv(FBI11_path, delimiter=',')
crime12_base = pd.read_csv(FBI12_path, delimiter=',')
crime13_base = pd.read_csv(FBI13_path, delimiter=',')
crime14_base = pd.read_csv(FBI14_path, delimiter=',')
crime15_base = pd.read_csv(FBI15_path, delimiter=',')
crime16_base = pd.read_csv(FBI16_path, delimiter=',')
crime17_base = pd.read_csv(FBI17_path, delimiter=',')
crime18_base = pd.read_csv(FBI18_path, delimiter=',')
crime19_base = pd.read_csv(FBI19_path, delimiter=',')

crime10_stats = crime10_base.fillna(0)
crime11_stats = crime11_base.fillna(0)
crime12_stats = crime12_base.fillna(0)
crime13_stats = crime13_base.fillna(0)
crime14_stats = crime14_base.fillna(0)
crime15_stats = crime15_base.fillna(0)
crime16_stats = crime16_base.fillna(0)
crime17_stats = crime17_base.fillna(0)
crime18_stats = crime18_base.fillna(0)
crime19_stats = crime19_base.fillna(0)

crime10_stats['Year'] = 2010
crime11_stats['Year'] = 2011
crime12_stats['Year'] = 2012
crime13_stats['Year'] = 2013
crime14_stats['Year'] = 2014
crime15_stats['Year'] = 2015
crime16_stats['Year'] = 2016
crime17_stats['Year'] = 2017
crime18_stats['Year'] = 2018
crime19_stats['Year'] = 2019
```

```python

crime10_complete = crime10_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime11_complete = crime11_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime12_complete = crime12_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime13_complete = crime13_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime14_complete = crime14_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime15_complete = crime15_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime16_complete = crime16_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime17_complete = crime17_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime18_complete = crime18_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime19_complete = crime19_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]


crime_stats_df = pd.concat([crime10_complete, 
                            crime11_complete, 
                            crime12_complete, 
                            crime13_complete, 
                            crime14_complete, 
                            crime15_complete, 
                            crime16_complete, 
                            crime17_complete, 
                            crime18_complete, 
                            crime19_complete], ignore_index=True, axis=0)
crime_stats_df
```

```python

crime_stats_df['Population'] = crime_stats_df['Population'].str.replace(',','')
crime_stats_df['Violent crime'] = crime_stats_df['Violent crime'].str.replace(',','')
crime_stats_df['Rape'] = crime_stats_df['Rape'].str.replace(',','')
crime_stats_df['Robbery'] = crime_stats_df['Robbery'].str.replace(',','')
crime_stats_df['Aggravated Assault'] = crime_stats_df['Aggravated Assault'].str.replace(',','')
crime_stats_df['Property Crime'] = crime_stats_df['Property Crime'].str.replace(',','')
crime_stats_df['Burglary'] = crime_stats_df['Burglary'].str.replace(',','')
crime_stats_df['Larceny-theft'] = crime_stats_df['Larceny-theft'].str.replace(',','')
crime_stats_df['Motor vehicle theft'] = crime_stats_df['Motor vehicle theft'].str.replace(',','')


crime_stats_df['Population'] = crime_stats_df['Population'].astype('float64')
crime_stats_df['Violent crime'] = crime_stats_df['Violent crime'].astype('float64')
crime_stats_df['Rape'] = crime_stats_df['Rape'].astype('float64')
crime_stats_df['Robbery'] = crime_stats_df['Robbery'].astype('float64')
crime_stats_df['Aggravated Assault'] = crime_stats_df['Aggravated Assault'].astype('float64')
crime_stats_df['Property Crime'] = crime_stats_df['Property Crime'].astype('float64')
crime_stats_df['Burglary'] = crime_stats_df['Burglary'].astype('float64')
crime_stats_df['Larceny-theft'] = crime_stats_df['Larceny-theft'].astype('float64')
crime_stats_df['Motor vehicle theft'] = crime_stats_df['Motor vehicle theft'].astype('float64')
crime_total = crime_stats_df.groupby(['City', 'Year']).sum()
crime_total.head()
```

```python
year_to_compare = 2019
cities_to_compare = ["Chicago", "Naperville", "Springfield", "Bolingbrook", "Park Ridge"]

filtered_data = crime_stats_df[(crime_stats_df['Year'] == year_to_compare) & (crime_stats_df['City'].isin(cities_to_compare))]

plt.figure(figsize=(10, 6))
plt.bar(filtered_data['City'], filtered_data['Property Crime'], color='skyblue')
plt.xlabel("Cities")
plt.ylabel("Property Crime Rate")
plt.title(f"Property Crime Rates in {year_to_compare} (Selected Cities)")
plt.grid(axis='y', linestyle='--', alpha=0.7)

#save graph
plt.savefig('output_data/img2.png')
```
