```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# import the state files
file_1 = Path("../Project_1/Resources/FBI_2010_illinois_amended.csv")
file_2 = Path("../Project_1/Resources/FBI_2011_illinois_amended.csv")
file_3 = Path("../Project_1/Resources/FBI_2012_illinois_amended.csv")
file_4 = Path("../Project_1/Resources/FBI_2013_illinois_amended.csv")
file_5 = Path("../Project_1/Resources/FBI_2014_illinois_amended.csv")
file_6 = Path("../Project_1/Resources/FBI_2015_illinois_amended.csv")
file_7 = Path("../Project_1/Resources/FBI_2016_illinois_amended.csv")
file_8 = Path("../Project_1/Resources/FBI_2017_illinois_amended.csv")
file_9 = Path("../Project_1/Resources/FBI_2018_illinois_amended.csv")
file_10 = Path("../Project_1/Resources/FBI_2019_illinois_amended.csv")

# Read the CSV files into a DataFrame
data_file_1 = pd.read_csv(file_1)
data_file_2 = pd.read_csv(file_2)
data_file_3 = pd.read_csv(file_3)
data_file_4 = pd.read_csv(file_4)
data_file_5 = pd.read_csv(file_5)
data_file_6 = pd.read_csv(file_6)
data_file_7 = pd.read_csv(file_7)
data_file_8 = pd.read_csv(file_8)
data_file_9 = pd.read_csv(file_9)
data_file_10 = pd.read_csv(file_10)

# Remove noisy columns
columns =['City','Population','Violent crime',	'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',	'Aggravated Assault',
         'Property Crime',	'Burglary',	'Larceny-theft', 'Motor vehicle theft',	'Arson']
data_file_1 =data_file_1[columns]   
data_file_2 =data_file_2[columns] 
data_file_3 =data_file_3[columns] 
data_file_4 =data_file_4[columns]       
data_file_5 =data_file_5[columns] 
data_file_6 =data_file_6[columns] 
data_file_7 =data_file_7[columns] 
data_file_8 =data_file_8[columns] 
data_file_9 =data_file_9[columns] 
data_file_10 =data_file_10[columns] 

# Print the resulting DataFrame
data_file_10.head()
```

```python
#Identify specific citys in different files (years)
All_Cities={}
city_list = ["Naperville","Chicago", "Springfield","Park Ridge","Bolingbrook"]
for City in city_list:
      
  row_2010  =data_file_1[data_file_1['City']==City]
  row_2010['Year'] ='2010'
  row_2011  =data_file_2[data_file_2['City']==City]
  row_2011['Year'] ='2011'
  row_2012  =data_file_3[data_file_3['City']==City]
  row_2012['Year'] ='2012'
  row_2013  =data_file_4[data_file_4['City']==City]
  row_2013['Year'] ='2013'
  row_2014  =data_file_5[data_file_5['City']==City]
  row_2014['Year'] ='2014'
  row_2015  =data_file_6[data_file_6['City']==City]
  row_2015['Year'] ='2015'
  row_2016  =data_file_7[data_file_7['City']==City]
  row_2016['Year'] ='2016'    
  row_2017  =data_file_8[data_file_8['City']==City]
  row_2017['Year'] ='2017'
  row_2018  =data_file_9[data_file_9['City']==City]
  row_2018['Year'] ='2018'
  row_2019  =data_file_10[data_file_10['City']==City]
  row_2019['Year'] ='2019'

  rows = [row_2010, row_2011, row_2012,row_2013, row_2014,row_2015, row_2016, row_2017,row_2018, row_2019]

  result = pd.concat(rows)

  All_Cities[City] = result
```

```python
# get the new 5 cities dataframes 
FBI_df_Naperville = All_Cities['Naperville']
FBI_df_Chicago = All_Cities['Chicago']
FBI_df_Springfield = All_Cities['Springfield']
FBI_df_Park_Ridge = All_Cities['Park Ridge']
FBI_df_Bolingbrook = All_Cities['Bolingbrook']

FBI_df_Naperville
```

```python
#convert 5 cities dataframes to csv files 
FBI_df_Naperville.to_csv('FBI_df_Naperville.csv', index=False)
FBI_df_Chicago.to_csv('FBI_df_Chicago.csv', index=False)
FBI_df_Springfield.to_csv('FBI_df_Springfield.csv', index=False)
FBI_df_Park_Ridge.to_csv('FBI_df_Park_Ridge.csv', index=False)
FBI_df_Bolingbrook.to_csv('FBI_df_Bolingbrook.csv', index=False)

# Specify the path to save the CSV file
file_path = ("../Project_1/Resources/FBI_df_Naperville.csv,FBI_df_Chicago.csv,FBI_df_Springfield.csv,FBI_df_Park_Ridge.csv,FBI_df_Bolingbrook.csv")

FBI_df_Naperville
```

```python
# combine the 5 cities dataframes into one df by "years"
# Create a list of DataFrames
df = [FBI_df_Naperville, FBI_df_Chicago, FBI_df_Springfield, FBI_df_Park_Ridge, FBI_df_Bolingbrook]

# Concatenate DataFrames based on the "years" column
combined_df = pd.concat(df, ignore_index=True)

# Now, 'combined_df' contains the data from all the DataFrames with a common "years" column.
combined_df  
```

```python
#filter cities data based on crime classfication and the years
# select columns by name

combined_filter_df = combined_df.filter(items=['City','Year', 'Rape','Robbery','Burglary','Larceny-theft','Motor vehicle theft'])

combined_filter_df.info
```

```python
# Step 1: Create a set of years for each city
years_by_city = {}
cities = combined_filter_df['City'].unique()

for city in cities:
    years_by_city[city] = set(combined_filter_df[combined_filter_df['City'] == city]['Year'])

# Step 2: Find the intersection of these sets to identify the common years
overlap_years = set.intersection(*years_by_city.values())

# 'overlap_years' now contains the common years among all cities.
print("Overlap Years:", sorted(list(overlap_years)))
```

```python
#creat the new df based on the overlap_years
FBI_new_city_df = combined_filter_df[combined_filter_df['Year'].isin(overlap_years)]
FBI_new_city_df.info
```

```python
FBI_new_city_df["Rape"] = [float(str(i).replace(",", "")) for i in FBI_new_city_df["Rape"]]
FBI_new_city_df["Robbery"] = [float(str(i).replace(",", "")) for i in FBI_new_city_df["Robbery"]]
FBI_new_city_df["Burglary"] = [float(str(i).replace(",", "")) for i in FBI_new_city_df["Burglary"]]
FBI_new_city_df["Larceny-theft"] = [float(str(i).replace(",", "")) for i in FBI_new_city_df["Larceny-theft"]]
FBI_new_city_df["Motor vehicle theft"] = [float(str(i).replace(",", "")) for i in FBI_new_city_df["Motor vehicle theft"]]
```
