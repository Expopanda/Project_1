```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

FBI15_path = 'Resources\FBI_2015_illinois_amended.csv'
FBI16_path = 'Resources\FBI_2016_illinois_amended.csv'
FBI17_path = 'Resources\FBI_2017_illinois_amended.csv'
FBI18_path = 'Resources\FBI_2018_illinois_amended.csv'
FBI19_path = 'Resources\FBI_2019_illinois_amended.csv'
```

```python
# these are the dataframes with all NaN values in tact. Keep these and don't amend them. 
# Use the "crime15_stats" dataframe
crime15_base = pd.read_csv(FBI15_path, delimiter=',')
crime16_base = pd.read_csv(FBI16_path, delimiter=',')
crime17_base = pd.read_csv(FBI17_path, delimiter=',')
crime18_base = pd.read_csv(FBI18_path, delimiter=',')
crime19_base = pd.read_csv(FBI19_path, delimiter=',')

crime15_stats = crime15_base.fillna(0)
crime16_stats = crime16_base.fillna(0)
crime17_stats = crime17_base.fillna(0)
crime18_stats = crime18_base.fillna(0)
crime19_stats = crime19_base.fillna(0)

crime15_stats['Year'] = 2015
crime16_stats['Year'] = 2016
crime17_stats['Year'] = 2017
crime18_stats['Year'] = 2018
crime19_stats['Year'] = 2019
crime15_stats.head()
```

```python
crime15_complete = crime15_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime16_complete = crime16_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime17_complete = crime17_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime18_complete = crime18_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
crime19_complete = crime19_stats[['City', 'Year', 'Population', 'Violent crime',
       'Murder and Non-Negligent Manslaughter', 'Rape', 'Robbery',
       'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
       'Motor vehicle theft', 'Arson']]
#crime15_complete.head()
#crime16_complete.head()
#crime17_complete.head()
#crime18_complete.head()
crime19_complete.head()
```

```python
crime_stats_df = pd.concat([crime15_complete, crime16_complete, crime17_complete, crime18_complete, crime19_complete], ignore_index=True, axis=0)
crime_stats_df
```
