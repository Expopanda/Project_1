```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data (replace this with your actual DataFrame)
data = {
    'years': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'male': [2506, 2691, 2566, 2416, 2078, 2062, 2608, 2112, 1989, 1697, 2001, 15713, 2001],
    'female': [389, 399, 357, 342, 320, 308, 304, 360, 409, 335, 349, 2503, 349],
}

offendersex = pd.DataFrame(data)

# Bar width
bar_width = 0.35

# Set up the bar positions
bar_positions_male = np.arange(len(offendersex))
bar_positions_female = bar_positions_male + bar_width

# Create the bar chart
plt.bar(bar_positions_male, offendersex['male'], width=bar_width, label='Male Crimes', color='blue')
plt.bar(bar_positions_female, offendersex['female'], width=bar_width, label='Female Crimes', color='pink')

# Adding labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.title('Crimes by Gender in Illinois')
plt.xticks(bar_positions_male + bar_width / 2, offendersex['years'])
plt.legend()

# Show the plot
plt.show()

plt.savefig('Crime_and_Gender.png')

```
